# E-Commerce Purchasing Intention Prediction

The dataset consists of features belonging to 12,330 sessions. 
The dataset was formed so that each session would belong to a different user in a 1-year period to avoid any tendency to a specific campaign, special day, user profile, or period.

Data of online shoppers' profiles were based on the sportswear website [Columbia](https://www.columbia.com.tr/) from Turkey, with screenshot as below. 

The data was donated to the [UCI machine learning website](http://archive.ics.uci.edu/ml/datasets/Online+Shoppers+Purchasing+Intention+Dataset).

![caption](Columbia.png "Columbia Shopping Site")

## Numeric features:

| Feature Name   | Description        | 
| :------------- | :----------: | 
|  Administrative| No. of pages visited by visitor on account management|
| Administrative_Duration  | Total time (secs) spent by visitor on account management related pages| 
| Informational | No. of pages visited by visitor about website, communication and address information of shopping site| 
| Informational_Duration | Total time (secs) spent by visitor on informational pages | 
| ProductRelated      | No. of pages visited by visitor on product-related pages| 
| ProductRelated_Duration  | Total time (secs) spent by visitor on product-related pages | 
| BounceRates| Average bounce rate of pages visited by visitor | 
| ExitRates| Average exit rate of pages visited by visitor | 
| PageValues| Average page value of pages visited by visitor| 
| SpecialDay| Closeness of date of site visit to a special day (between 0 to 1)| 

## Categorical Features:

| Feature Name   | Type         | 
| :------------- | :----------: | 
|  Month| Month of visit date  |
| VisitorType    | Visitor Type (New visitor, Returning Visitor, Others)| 
| Weekend| True/False | 
| Revenue| Class label indicating whether visit is finalised with a transaction | 

## Masked Features (Categorical):

| Feature Name   | Type         | 
| :------------- | :----------: | 
| OperatingSystems   | Operating System of visitor | 
| Browser| Browser of visitor | 
| Region  | Geographic Region where visitor started session | 
| TrafficType   | Traffic source (e.g. banner, SMS, direct) |

### 3.1.2 Cross Validation

```python
#create various models to check scores using above function

logreg = LogisticRegression(random_state=42,n_jobs=-1)
rfc = RandomForestClassifier(n_estimators=10,random_state=42,n_jobs=-1)
gbc = GradientBoostingClassifier(random_state=42)
cbc = CatBoostClassifier(random_state=42,used_ram_limit='10gb',verbose=False)
lgbmc = LGBMClassifier(random_state=42)
xgbc = XGBClassifier(n_jobs=-1,random_state=42)
linSVC = LinearSVC(random_state=42,dual=False)
svc = SVC(kernel="rbf", probability=True,random_state=42,gamma='auto')
mlp10 = MLPClassifier(random_state=42,hidden_layer_sizes=(10, ))


models = [logreg,rfc, gbc, cbc, lgbmc,xgbc, linSVC, svc, mlp10]
model_names = ['Logistic Regression','Random Forest','Gradient Boosting','CatBoost','LightGBM','XGBoost','Linear SVM', 'Gaussian SVM','MLP 10']
```

```python
#define a function that helps to compare cross-validated scores of models

#for imbalanced class problem
def models_score_imb(models,X_train,y_train):
    scores_dict = {}
    
    print('Cross-validation.....start!')
    print('-'*50)
    
    for i,model in enumerate(models):
        
        print('Model',i+1,':',model_names[i])
        
        skf = StratifiedKFold(n_splits=10,random_state=42)
        
        scores = []

        for k, (train_index, test_index) in enumerate(skf.split(X_train,y_train)):
            
            print('Fold no',k+1,'in progress....')
            
            Xfold_train, Xfold_test = X_train.iloc[train_index,:], X_train.iloc[test_index,:]
            yfold_train, yfold_test = y_train.iloc[train_index], y_train.iloc[test_index]

            
            #ADASYN oversampling tool
            upsam = ADASYN(sampling_strategy='minority', random_state=42,
              n_jobs=-1)
            X_upsam, y_upsam = upsam.fit_sample(Xfold_train, yfold_train)

            X_upsam, y_upsam = np.array(X_upsam),np.array(y_upsam)
            Xfold_test = np.array(Xfold_test)
            
            model.fit(X_upsam, y_upsam)
            yfold_pred = model.predict(Xfold_test)
            score = roc_auc_score(yfold_test,yfold_pred)
            #score = accuracy_score(yfold_test,yfold_pred)
            scores.append(score)
        
        scores_dict[model_names[i]] = scores
        print('Done!')
        print('-'*50)
        
    df = pd.DataFrame(scores_dict)
    df.columns = model_names
    
    return df
```

```python
models_scoring_imb = models_score_imb(models,Xs_train,y_train)
```

```python
#define a function to show various classification metrics
def classification_metrics_imb(models,X_train,Xs_test,y_train,y_test):
    
    scores_dict = {}
    
    for i, model in enumerate(models):
        start_time = time.clock()
        
        print(model_names[i])
        print()

        
        scores = []
        
        #ADASYN oversampling tool
        upsam = ADASYN(sampling_strategy='minority', random_state=42,n_jobs=-1)
        X_upsam, y_upsam = upsam.fit_sample(X_train, y_train)
        
        X_upsam, y_upsam = np.array(X_upsam),np.array(y_upsam)
        XsA_test = np.array(Xs_test)
        
        model = model.fit(X_upsam,y_upsam)
        y_pred = model.predict(XsA_test)
        time_diff = time.clock() - start_time
        print(" %s seconds taken for model to run...." %time_diff)
        print()
#         try:
#             y_pred_proba = model.predict_proba(X_test)
#         except:
#             pass
            

        print(pd.DataFrame(confusion_matrix(y_test,y_pred),columns=['Predicted Non Purchase', 'Predicted Purchase'],
                     index=['Actual Non Purchase','Actual Purchase']))
        print()
        print(classification_report(y_test,y_pred))
        scores.append(accuracy_score(y_test,y_pred))
        scores.append(roc_auc_score(y_test,y_pred))
        scores.append(precision_score(y_test,y_pred))
        scores.append(recall_score(y_test,y_pred))
        scores.append(f1_score(y_test,y_pred))
        scores.append(time_diff)
#        scores.append(roc_auc_score(y_test,y_pred_proba[:,1]))
        
#         try:
#             print('ROC AUC score:',roc_auc_score(y_test,y_pred_proba[:,1]))
#         except:
#              pass
        
        print('-'*50)
        
        scores_dict[model_names[i]] = scores

        
    df = pd.DataFrame(scores_dict)
    df.columns =  model_names
    df.index = ['Accuracy','ROC AUC','Precision','Recall','F1 Score','Time Taken']
    
    return df
```
